```r
# read csv file 
dataset <- read.csv ("h2d.csv", na.strings="",stringsAsFactors=FALSE)
```

```r
head (dataset,n=20) # print top 5 rows 
```

install.packages("ggplot2")
library(ggplot2)


library(psych)
pairs.panels(dataset)

```r
plot(dataset$KB/1000, dataset$BW*1,xlab = "Array size [MB]", ylab = "Bandwidth [MB/s]", pch=21, main="H2D Bandwith vs Array size")
```

```r
library(psych)
pairs.panels(dataset, 
             method = "pearson", # correlation method
             hist.col = "#00AFBB",
             density = TRUE,  # show density plots
             ellipses = TRUE # show correlation ellipses
)
```

```r
# model : H2D

mod1<-lm(BW~KB,data=dataset)
summary(mod1) # display summary 
new<-data.frame(KB=(58982.4)) # predication value 
res<-predict(mod1,newdata=new) # predication result 
cat('BW [kbps]:',res)

cat('\nH2D time [ms]:',(58982.4/res)*1000)
```
