```r
# read csv file 
dataset <- read.csv ("d2h.csv", na.strings="",stringsAsFactors=FALSE)
```

```r
head (dataset,n=20) # print top 5 rows 
```

```r
plot(dataset$KB/1000, dataset$BW1,xlab = "Array size [MB]", ylab = "Bandwidth [MB/s]", pch=21, main="D2H Bandwith vs Array size")
```

```r
library(psych)
pairs.panels(dataset, 
             method = "pearson", # correlation method
             hist.col = "#00AFBB",
             density = TRUE,  # show density plots
             ellipses = TRUE # show correlation ellipses
)
```

```r

# model : D2H

mod1<-lm(BW1~KB,data=dataset)
summary(mod1) # display summary 
new<-data.frame(KB=(58982.4)) # predication value 
res<-predict(mod1,newdata=new) # predication result 
cat('BW [kbps]:',res)

cat('\nD2H time [ms]:',(58982.4/res)*1000)
```
